# 概日時計システム

* Goodwin model
* $x$: mRNA の濃度、$y$: たんぱく質の濃度、$k_x, k_y$: $x,y$の合成速度、$d_x, d_y$: $x,y$の分解速度 
$$\frac{dx}{dt} = \frac{k_x K}{K+y}-d_x, \quad 
\frac{dy}{dt} = k_y x - d_y
$$ 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp 
kx = 1
ky = 3
dx = 0.3
dy = 0.3
K = 1
x = 10
y = 10
def deriv(t, z, kx,ky,dx,dy,K):
    x, y = z
    return [kx*K/(K+y)-dx, ky*x-dy]
tspan = [0, 300]
teval = np.linspace(0, 100, 1000)
z0 = [x,y]
sol = solve_ivp(deriv, t_span=tspan, y0=z0, t_eval=teval, args=(kx,ky,dx,dy,K,))
plt.plot(sol.t, sol.y[0])
plt.plot(sol.t, sol.y[1])
```
